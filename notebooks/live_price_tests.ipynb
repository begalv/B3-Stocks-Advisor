```python
from yflive import QuoteStreamer
```

```python
qs = QuoteStreamer()
```

```python
qs.subscribe(["BPAC11.SA", "PETR4.SA"])
```

```python
class Stock:

    def __init__(self, ticker):
        self.ticker = ticker
        self.price = 0
        
    def set_price(self, quote):
        if quote.identifier == self.ticker:
            self.price = quote.price
```

```python
bpac11 = Stock("BPAC11.SA")
```

```python
petr4 = Stock("PETR4.SA")
```

```python
def set_prices(stocks, quote):
    for stock in stocks:
        stock.set_price(quote)
```

```python
qs.on_quote = lambda qs, q: set_prices([bpac11, petr4], q)
qs.start(should_thread=True)
```

```python
print(bpac11.price)
```

```python
print(petr4.price) 
```
