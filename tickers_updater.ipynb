```python
import requests
import pandas as pd
import json
from pathlib import Path
from delfos.market.session import Session
import requests
```

```python
session = Session(period=6)
```

```python
with open(Path("./delfos/data/market/tickers.json"), encoding="utf8") as tickers_file:
    TICKERS_DICT = json.load(tickers_file) #dicionário com todas as informações fixas das ações listadas na B3
    
FUNDAMENTUS_URL = "https://fundamentus.com.br/detalhes.php?papel="
REQUESTS_HEADER = {"User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36", 'Accept-Encoding': 'gzip, deflate', 'Accept': '*/*', 'Connection': 'keep-alive'}
```

```python
#PRINCIPAIS AÇÕES DA BOLSA (269 ATÉ O MOMENTO)
tickers_list_url = "https://www.mobills.com.br/blog/investimentos/cnpj-das-empresas-listadas-na-b3/" #trocar?
```

```python
possible_new_tickers_df = pd.read_html(requests.get(tickers_list_url).text)[0]
possible_new_tickers_df = possible_new_tickers_df.drop_duplicates(keep="last")
possible_new_tickers_df["TICKER"] = possible_new_tickers_df["TICKER"].str.strip()
```

```python
#TICKERS QUE ESTÃO NO URL MAS NÃO ESTÃO NO SISTEMA
new_tickers = []
for possible_ticker in possible_new_tickers_df["TICKER"]:
    if possible_ticker not in session.stocks:
        new_tickers.append(possible_ticker)
        
len(new_tickers)
```

```python
#PEGAR DADOS QUE FALTAM DAS NOVAS AÇÕES E GRAVAR NO ARQUIVO JSON

stocks_not_found = 0
for ticker in new_tickers:
    try:
        requested_data = requests.get(FUNDAMENTUS_URL + ticker, headers=REQUESTS_HEADER ) #chamada GET ao site da Fundamentus
        html_tables_list = pd.read_html(requested_data.text)

        session_name = html_tables_list[0][1][2] #nome da companhia

        cnpj = possible_new_tickers_df["CNPJ"][possible_new_tickers_df["TICKER"] == ticker].iloc[0]
        name = possible_new_tickers_df["RAZÃO SOCIAL"][possible_new_tickers_df["TICKER"] == ticker].iloc[0]

        sector = html_tables_list[0][1][3] #setor da companhia
        sub_sector = html_tables_list[0][1][4] #sub setor da companhia
        segment = html_tables_list[0][1][4] #segmento da companhia

        ticker_dict = {ticker: {'Nome de pregão':session_name, "CNPJ": cnpj, "Nome": name, "Setor": sector, "Sub Setor":sub_sector, "Segmento": segment}}

        TICKERS_DICT.update(ticker_dict)
        
        with open(Path("./delfos/data/market/tickers.json"), "w", encoding="utf8") as json_file:
            json.dump(TICKERS_DICT, json_file, indent=4, separators=(',',': '))
    
    except Exception as e:
        stocks_not_found +=1
        pass #O PAPEL NÃO ESTÁ NO SITE FUNDAMENTUS, PORTANTO NÃO APRESENTA TODAS AS INFOS NECESSÁRIAS PARA SER ADICIONADO AO JSON

print(stocks_not_found)
```

```python
#TICKERS QUE ESTÃO NO SISTEMA MAS NÃO ESTÃO NO URL
possible_old_tickers = []
for ticker in session.stocks:
    if ticker not in possible_new_tickers_df["TICKER"].tolist():
        possible_old_tickers.append(ticker)
        
len(possible_old_tickers)
```
